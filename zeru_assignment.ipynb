```python
import pandas as pd
import numpy as np
```

## Understanding of Dataset

```python
dataframe=pd.read_csv('user_wallet1.csv')
dataframe.head()
```

```python
column_name=dataframe.columns
print(column_name)
```

```python
names=dataframe.drop_duplicates(subset=['type'],keep='first')

names
```

# Dropping useless columns

```python
dataframe=dataframe.drop(columns=['poolId','toId','callerId', 'useATokens', 'repayerId',
       'liquidatorId','principalReserveId','collateralReserveId'],axis=1)
```

```python
dataframe.head()
```

```python
asset_symbol=dataframe['userId'].nunique()
print(asset_symbol)
```

```python
amount_cols = ['amount', 'assetPriceUSD', 'borrowRate', 'variableTokenDebt',
               'stableTokenDebt', 'collateralAmount', 'collateralAssetPriceUSD',
               'principalAmount', 'borrowAssetPriceUSD']
```

Making columns to numeric

```python
for col in amount_cols:
    if col in dataframe.columns:
        dataframe[col] = pd.to_numeric(dataframe[col], errors='coerce')
```

calculating USD value

```python
dataframe["usd_value"] = dataframe["amount"] * dataframe["assetPriceUSD"]
```

```python
dataframe.head()
```

Now aggregating a wallet on userId and calculating total deposit,total borrow,total repay and other attributes

```python
agg = dataframe.groupby("userId").agg(
    total_Deposit_usd=pd.NamedAgg(column="usd_value", aggfunc=lambda x: x[dataframe['type'] == 'Deposit'].sum()),
    total_Borrow_usd=pd.NamedAgg(column="usd_value", aggfunc=lambda x: x[dataframe['type'] == 'Borrow'].sum()),
    total_Repay_usd=pd.NamedAgg(column="usd_value", aggfunc=lambda x: x[dataframe['type'] == 'Repay'].sum()),
    total_liquification_usd=pd.NamedAgg(column="usd_value", aggfunc=lambda x: x[dataframe['type'] == 'LiquidationCall'].sum()),
    total_redeemunderlying_usd=pd.NamedAgg(column="usd_value", aggfunc=lambda x: x[dataframe['type'] == 'RedeemUnderlying'].sum()),
    num_Deposits=pd.NamedAgg(column="type", aggfunc=lambda x: (x == 'Deposit').sum()),
    num_Borrows=pd.NamedAgg(column="type", aggfunc=lambda x: (x == 'Borrow').sum()),
    num_Repays=pd.NamedAgg(column="type", aggfunc=lambda x: (x == 'Repay').sum()),
    avg_Borrow_rate=pd.NamedAgg(column="borrowRate", aggfunc="mean"),
    total_collateral_usd=pd.NamedAgg(column="collateralAmount", aggfunc=lambda x: (x * dataframe["collateralAssetPriceUSD"]).sum()),
    num_unique_assets=pd.NamedAgg(column="assetSymbol", aggfunc=lambda x: x.nunique())
).fillna(0)
```

```python
agg.head()
```

saving aggregated dataframe in csv formate


```python
agg.reset_index(inplace=True)
agg.to_csv("wallet_features.csv", index=False)
```

# Performing Normalisation

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
agg.iloc[:,1:]=sc.fit_transform(agg.iloc[:,1:])
```

```python
agg.head()
```

```python
numeric_df=agg.drop(columns=["userId","total_liquification_usd"],axis=1)
```

 Creating a heuristic credit_score to make as dependent variable or label o dataset

```python
agg["credit_score"] = (
    0.4 * numeric_df["total_Repay_usd"] +
    # 0.1 * numeric_df["total_redeemunderlying_usd"] +
    0.2 * numeric_df["total_Deposit_usd"] +
    0.15 * numeric_df["num_Deposits"] +
    0.15 * (1-numeric_df["avg_Borrow_rate"]) +  # low borrow rate is good
    0.1 * numeric_df["total_collateral_usd"]
) * 2000
```

clipping score from 0 to 1000 as problem statement mentions it

```python
agg['credit_score']=agg['credit_score'].clip(0,1000)
```

```python
agg.describe()
```

removing total_liquification_usd,total_redeemunderlying_usd as they have very less values and can cause underfitting for our model

```python
agg=agg.drop(columns=['total_liquification_usd','total_redeemunderlying_usd'],axis=1)
```

```python
agg.head()
```

# Separating independent and dependent attributes

```python
X=agg.iloc[:,1:-1].values
y=agg.iloc[:,-1].values
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

# XGBoost Regressor

```python
from xgboost import XGBRegressor

regressor1=XGBRegressor()

regressor1.fit(X_train,y_train)
pred_train=regressor1.predict(X_train)
pred_test=regressor1.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_train,pred_train))
print(r2_score(y_test,pred_test))
```

# linear regression

```python
from sklearn.linear_model import LinearRegression
regressor2=LinearRegression(fit_intercept=True)
regressor2.fit(X_train,y_train)
pred_train_linear=regressor2.predict(X_train)
pred_test_linear=regressor2.predict(X_test)
```

```python
print(r2_score(y_train,pred_train_linear))
print(r2_score(y_test,pred_test_linear))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
regressor3=DecisionTreeRegressor()
regressor3.fit(X_train,y_train)
pred_train_DTR=regressor3.predict(X_train)
pred_test_DTR=regressor3.predict(X_test)
```

```python
print(r2_score(y_train,pred_train_DTR))
print(r2_score(y_test,pred_test_DTR))
```

# Random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
regressor4=RandomForestRegressor(n_estimators=7)
regressor4.fit(X_train,y_train)
pred_train_RFR=regressor4.predict(X_train)
pred_test_RFR=regressor4.predict(X_test)
```

```python
print(r2_score(y_train,pred_train_RFR))
print(r2_score(y_test,pred_test_RFR))
```

# K fold cross validation

```python
regressor=[regressor1,regressor2,regressor3,regressor4]

for regress in regressor:
  from sklearn.model_selection import cross_val_score
  accuracies1=cross_val_score(estimator=regress,X=X_train,y=y_train,cv=10)
  print(accuracies1.mean(),regress)
```

# Classification on basics of score

```python
def classify(prediction):
  if prediction>=500:
    print("Reliable")
  elif 500>prediction>=300:
    print("Moderate")
  else:
    print("Unreliable")
```
